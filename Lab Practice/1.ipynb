# Take grayscale image of size 512x512 and perform the following operations

## a. Decrease its spatial resolution by half every time and observe its change when displaying in the same window size

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
```

```python
original_img = mpimg.imread("bridge.jpg")
img = original_img

imgs = []
for i in range(4):
  imgs.append(img)
  height = img.shape[0] // 2
  width = img.shape[1] // 2
  img = cv2.resize(img, (width, height), interpolation=cv2.INTER_LINEAR)
imgs.append(img)
plt.figure(figsize=(16, 12))
for i, img in enumerate(imgs):
  plt.subplot(1, len(imgs), i + 1)
  plt.imshow(img, cmap='gray')
  plt.title(f"{img.shape[1]}x{img.shape[0]}")
  plt.axis('off')
plt.tight_layout()
plt.show()
```

# Decrease it intensity level resolution by one bit up to reach its binary format observe its change when displaying in the same window size


```python
img = original_img

imgs = []

for bits in range(0, 8):
  levels = 2 ** bits
  temp = np.floor(img / levels)
  imgs.append((temp, bits))

plt.figure(figsize=(16, 8))
for i, (img, bits) in enumerate(imgs):
  plt.subplot(2, 4, i + 1)
  plt.imshow(img, cmap='gray')
  plt.title(f"{bits} bits ({2**bits} levels)")
  plt.axis('off')

plt.tight_layout()
plt.show()
```
