**Import Necessary Modules**

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
import numpy as np
```

```python
original_img = cv2.imread('Lenna-gray.bmp')
normalized_img = original_img / 255.0
plt.imshow(original_img, cmap = 'gray')
plt.show()
```

#(a)Perform the brightness enhancement of a specific range of gray levels and observe its result

```python
plt.figure(figsize=(16, 12))
plt.subplot(1, 2, 1)
plt.imshow(original_img, cmap = 'gray')
plt.title('Original Image')

lower_limit = 100
upper_limit = 200

enhanced_img = original_img
enhanced_img[(enhanced_img >= lower_limit) & (enhanced_img <= upper_limit)] += 200
enhanced_img[(enhanced_img < lower_limit) | (enhanced_img > upper_limit)] -= 200

plt.subplot(1, 2, 2)
plt.imshow(enhanced_img, cmap = 'gray')
plt.title('Enhanced Image')
plt.show()
```

# (b) Differentiate the results of power law transform and inverse logarithmic transform

##Function for Power Law Transformation

```python
def power_law_transform(image, gamma):
  return np.clip((image ** gamma) * 255, 0, 255)

def inverse_log_transform(image):
  return np.clip((np.exp(image) - 1) * (255 / (np.exp(1) - 1)), 0, 255)

gammas, subplot_number = [1, 0.6, 0.4, 0.3], 1
plt.figure(figsize = (14, 17))

for gamma in (gammas):
    transformed_image = power_law_transform(original_img, gamma)
    plt.subplot(2, 2, subplot_number)
    plt.imshow(transformed_image, cmap = 'gray')
    plt.title(f"Power Law Transformation with Î³ = {gamma}")
    subplot_number += 1

plt.show()
```
