## Import necessary modules

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# Load image

```python
img = mpimg.imread('/content/bridge.jpg')
plt.figure(figsize=(6, 6))
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

# Extract all the bit planes

```python
bit_planes = []

for i in range(8):
  bit_plane = img & (2 ** (7 - i))
  bit_planes.append(bit_plane)

plt.figure(figsize=(6, 6))

plt.subplot(3, 3, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

for i in range(8):
    plt.subplot(3, 3, i + 2)
    plt.title(f'Bit Plane {(8 - i)}')
    plt.imshow(bit_planes[i], cmap='gray')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Combine bit planes from MSB to LSB

```python
combined_imgs = []
combined_img = np.zeros_like(img, dtype=np.uint8)

for plane in bit_planes:
    combined_img |= plane
    combined_imgs.append(combined_img.copy())

plt.figure(figsize=(8, 8))

plt.subplot(3, 3, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

for i in range(8):
    plt.subplot(3, 3, i + 2)
    plt.title(f'Bit Plane {(8 - i)} to 8')
    plt.imshow(combined_imgs[i], cmap='gray')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Combine bit planes from LSB to MSB

```python
combined_imgs = []
combined_img = np.zeros_like(img, dtype=np.uint8)

bit_planes = bit_planes[::-1]

for plane in bit_planes:
    combined_img |= plane
    combined_imgs.append(combined_img.copy())

plt.figure(figsize=(8, 8))

plt.subplot(3, 3, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

for i in range(8):
    plt.subplot(3, 3, i + 2)
    plt.title(f'Bit Plane 1 to {(i + 1)}')
    plt.imshow(combined_imgs[i], cmap='gray')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
