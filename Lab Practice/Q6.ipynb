# 6. Take a binary image and a structuring element to perform the following morphological operations â€“


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("wirebond.tif", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.axis("off")
plt.show()
```

```python
print(image.shape)
print(image[20:40, 23:50])

kernel_5 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
kernel_10 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10, 10))
kernel_20 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (20, 20))

erod_5 = cv2.erode(image, kernel_5, iterations=1)
erod_10 = cv2.erode(image, kernel_10, iterations=1)
erod_20 = cv2.erode(image, kernel_20, iterations=1)

fig, ax = plt.subplots(1, 4, figsize=(15, 5))

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(erod_5, cmap='gray')
ax[1].set_title("Eroded Image (5x5)")
ax[1].axis("off")

ax[2].imshow(erod_10, cmap='gray')
ax[2].set_title("Eroded Image (10x10)")
ax[2].axis("off")

ax[3].imshow(erod_20, cmap='gray')
ax[3].set_title("Eroded Image (20x20)")
ax[3].axis("off")

plt.show()
```

```python
# Delaation Operation radius 5, 10, 20
image = cv2.imread("shadedText.tif", cv2.IMREAD_GRAYSCALE)

# Thresholding image
_, image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)


# bitwise not
image = cv2.bitwise_not(image)

kernel_5 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
kernel_10 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10, 10))
kernel_20 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (20, 20))

dil_5 = cv2.dilate(image, kernel_5, iterations=1)
dil_10 = cv2.dilate(image, kernel_10, iterations=1)
dil_20 = cv2.dilate(image, kernel_20, iterations=1)

fig, ax = plt.subplots(1, 4, figsize=(15, 5))

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(dil_5, cmap='gray')
ax[1].set_title("Dilated Image (5x5)")
ax[1].axis("off")

ax[2].imshow(dil_10, cmap='gray')
ax[2].set_title("Dilated Image (10x10)")
ax[2].axis("off")

ax[3].imshow(dil_20, cmap='gray')
ax[3].set_title("Dilated Image (20x20)")
ax[3].axis("off")

plt.show()
```
