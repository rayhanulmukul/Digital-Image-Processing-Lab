# Import Necessary Module

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Take grayscale image of size 512x512 and perform the following operations -


```python
image = cv2.imread('flower.tif', cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (512, 512))

plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

# Decrease its spatial resolution by half every time and observe its change when displaying in the same window size


```python
imgs = []
tmp_img = image.copy()

for i in range(6):
  imgs.append(tmp_img);
  tmp_img = tmp_img[::2, ::2]


plt.figure(figsize=(10, 10))
for i in range(6):
  plt.subplot(2, 3, i + 1)
  plt.imshow(imgs[i], cmap='gray')
  plt.title(f"{imgs[i].shape[0]} x {imgs[i].shape[1]}")
  plt.axis('off')

plt.tight_layout()
plt.show()
```

# Decrease it intensity level resolution by one bit up to reach its binary format observe its change when displaying in the same window size

```python
image = cv2.imread('ctskull.tif', cv2.IMREAD_GRAYSCALE)

imgs = []

for bits in range(0, 8):
  level = 2 ** bits # 2^bits
  # print(f"Level: {level}")
  # print(imgs)
  imgs.append(image // level)

plt.figure(figsize=(10, 10))
for i in range(8):
  plt.subplot(2, 4, i + 1)
  plt.imshow(imgs[i], cmap='gray')
  plt.title(f"{8 - i} bits\n ({2**(8 - i)} leves)")
  plt.axis('off')

plt.tight_layout()
plt.show()
```

# Illustrate the histogram of the image and make single threshold segmentation observed from the histogram

```python
print(image.shape)
print(image.dtype)

histogram = np.zeros(256, dtype=int)

for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    histogram[image[i][j]] += 1

# Here Threshold Value T = 100
thresholded_img = image.copy()
thresholded_img[thresholded_img < 100] = 0
thresholded_img[thresholded_img >= 100] = 255

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.bar(range(256), histogram, color='gray')
plt.title('Histogram')
plt.xlabel("Gray Level(0-255)")
plt.ylabel("Number of Pixels")

plt.subplot(1, 2, 2)
plt.imshow(thresholded_img, cmap='gray')
plt.title('Thresholded Image')
plt.axis('off')

plt.show()
```
