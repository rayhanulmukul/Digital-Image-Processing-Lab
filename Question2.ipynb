# Import Necessary Module

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Take grayscale image of size 512x512 and perform the following operations â€“

```python
image = cv2.imread('kidney.tif', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(8, 4))
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.show()
```

# Perform the brightness enhancement of a specific range of gray levels and observe its result

```python
lower = 130
upper = 230

image1 = image.copy()
image2 = image.copy()

for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    if lower <= image[i][j] <= upper:
      image1[i][j] = 255
      image2[i][j] = 255
    else:
      image1[i][j] = 0

plt.figure(figsize=(16, 4))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Brightness Enhancement')
plt.imshow(image1, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Inverted Brightness Enhancement')
plt.imshow(image2, cmap='gray')

plt.show()
```

# Differentiate the results of power law transform and inverse logarithmic transform


**Power Law Transform:**
$\qquad s = c \times r^\gamma$, where $c = 1$

**Inverse Logarithmic Transform:**
$\qquad s = (e^{r/c} - 1)$, where $c = 255 / \log(1 + 255)$

```python
image = cv2.imread('runway.tif', cv2.IMREAD_GRAYSCALE)

powLaw_img = np.power(image / 255.0, 4.0) * 255.0

c = 255 / np.log(1.0 + 255)
invLog_img = np.exp(image / c) - 1

plt.figure(figsize=(16, 4))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Power Law Transform')
plt.imshow(powLaw_img, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Inverse Logarithmic Transform')
plt.imshow(invLog_img, cmap='gray')

plt.show()
```

# Find the difference image between original and the image obtained by last three bits (MSB)

```python
image = cv2.imread('dollar.tif', cv2.IMREAD_GRAYSCALE)

lastThreeBit = image & 7

reconstructedImg = lastThreeBit << 5

differenceImg = cv2.absdiff(image, reconstructedImg)

plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Difference Image')
plt.imshow(differenceImg, cmap='gray')

plt.show()
```

```python
image = cv2.imread('dollar.tif', cv2.IMREAD_GRAYSCALE)

lastThreeBit = image & 0b11100000

plt.figure(figsize=(20, 8))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Last Three Bits (MSB)')
plt.imshow(lastThreeBit, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Difference Image')
plt.imshow(image - lastThreeBit, cmap='gray')

plt.show()
```
